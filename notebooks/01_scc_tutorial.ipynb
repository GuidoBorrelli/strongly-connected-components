# SCC Tutorial

Load a curated graph, run all registered SCC algorithms, and compare their component partitions.

```python
import networkx as nx
from matplotlib import pyplot as plt

from algorithms.registry import list_algorithms, run_algorithm
from graph_io.examples import load_example_graph
```

```python
graph = load_example_graph("branched_scc")
results = {
    algorithm.name: run_algorithm(algorithm.key, graph)
    for algorithm in list_algorithms(category="scc")
}

for name, result in results.items():
    print(name, result.component_count, result.component_map)
```

```python
pos = nx.spring_layout(graph, seed=7)
nx.draw_networkx(graph, pos=pos)
plt.title("Curated SCC example")
plt.show()
```
