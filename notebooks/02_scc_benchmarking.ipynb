# SCC Benchmarking

Run a tiny benchmark suite, export the results, and inspect the summary table.

```python
from pathlib import Path

import pandas as pd

from benchmarks.exporters import export_benchmark_results
from benchmarks.runner import run_benchmark_suite
from benchmarks.suites import build_default_scc_suite
```

```python
cases = build_default_scc_suite(node_counts=(20, 40))
run_records, summary_records, metadata = run_benchmark_suite(
    cases,
    repeat_count=3,
    suite_name="notebook-demo",
    base_seed=100,
)
export_dir = export_benchmark_results(
    run_records,
    summary_records,
    metadata,
    output_root=Path("results"),
)
export_dir
```

```python
summary = pd.read_csv(export_dir / "summary.csv")
summary[[
    "case_key",
    "algorithm_name",
    "mean_runtime_milliseconds",
    "variance_runtime_milliseconds",
]]
```
